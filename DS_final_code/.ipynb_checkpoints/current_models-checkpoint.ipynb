## Set binary threshold anything above 3 is considered a success (good rating from the star rating)

```python
def convert(v):
    updated = []
    for i in v:
        if i > 3:
            updated.append(1)
        else:
            updated.append(0)
    return updated
```

```python
binary_from_threshold_asian = convert(star_ratings_a)
```

```python
binary_from_threshold_eu = convert(star_ratings_eu)
```

```python
3#is it reading .5 as contiuous 
##will round down to see if thats the case
###did not like the floats,, BUT why didnt it like the .5 
```

```python
def round_down(t):
    new_rating = []
    for i in t:
        if i == 1.5:
            i = 1.0
            new_rating.append(i)
        elif i == 2.5:
            i = 2.0
            new_rating.append(i)
        elif i == 3.5:
            i = 3.0
            new_rating.append(i)
        elif i == 4.5:
            i = 4.0
            new_rating.append(i)
        else:
            new_rating.append(i)
    return new_rating
```

```python
holds = round_down(star_ratings_a)
```

```python
type(holds)
```

```python
X_train, X_test, y_train, y_test = train_test_split(logistic_x_features_a1,holds, test_size=0.2)
```

```python
#model.fit(X_train, y_train)
train_multi = my_logistic_regression.fit(X_train, y_train)
```

```python
train_multi.score(X_train, y_train)
```
