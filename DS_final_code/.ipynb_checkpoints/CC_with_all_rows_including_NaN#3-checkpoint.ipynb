```python
def convert_float_to_str_afterdict(f):
    all_strs3 = []
    for i in f:
        
        if isinstance(i,float):
            #note even though you put the variable in the list you only get the new converted value
            new = str(i)
            all_strs3.append(new)
        else:
            all_strs3.append(i)
    return all_strs3
```

```python
def convert_nan_to_bool(q):
    temp = []
    for i in q:
        if i == 'nan':
            i = False
            #s = str(i)
            temp.append(i)
        elif i == 'None':
            i =  False
            #s = str(i)
            temp.append(i)
        else:
            temp.append(i)
    return temp
all_strings = convert_nan_to_bool(strings_now)
```
