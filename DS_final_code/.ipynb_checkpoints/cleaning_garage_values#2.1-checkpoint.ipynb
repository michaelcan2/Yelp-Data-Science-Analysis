```python
def convert_float_to_str_afterdict(f):
    all_strs3 = []
    for i in f:
        
        if isinstance(i,float):
            #note even though you put the variable in the list you only get the new converted value
            new = str(i)
            all_strs3.append(new)
        else:
            all_strs3.append(i)
    return all_strs3
```

```python
def convert_nan_to_dict(q):
    temp = []
    for i in q:
        if i == 'nan':
            i = {'garage' : False}
            #s = str(i)
            temp.append(i)
        elif i == 'None':
            i = {'garage' : False}
            #s = str(i)
            temp.append(i)
        else:
            temp.append(i)
    return temp
all_nans_now_dicts = convert_nan_to_dict(lists_of_strings_in_dict)
```

```python
def remove_brackets_(f):
    all_strs_list5 = []
    for i in f:
        
        if isinstance(i,str):
            sub = i.strip('{}') 
            all_strs_list5.append(sub)
            #all_strs5.append(new)
        else:
            
            all_strs_list5.append(i)
    return all_strs_list5

all_str_remove_brackets = remove_brackets_(all_nans_now_dicts) 
```

```python
def convert_str_to_lists(f):
    all_strs_list6 = []
    for i in f:
        
        if isinstance(i,str):
            sub = i.split(', ') 
            all_strs_list6.append(sub)
          
        else:
            
            all_strs_list6.append(i)
    return all_strs_list6

all_str_now_list_had_brackets_removed = convert_str_to_lists(all_str_remove_brackets)  
```

```python
def convert_str_to_dict(f):
    all_strs4 = []
    for i in f:
        if isinstance(i,str):
            new = ast.literal_eval(i)
            all_strs4.append(new)
        else:
            print(i)
            all_strs4.append(i)
    return all_strs4
```
